```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler 
from sklearn import datasets 
boston = datasets.load_boston()
boston_df = pd.DataFrame(boston.data,
columns=boston.feature_names)
targets = boston.target
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold

# Normalize the variables
scaler = StandardScaler()
features_norm = scaler.fit_transform(boston_df)

# Split the data
sss = StratifiedKFold(n_splits=5, random_state=None, shuffle=False)
X_train, X_test, y_train, y_test = train_test_split(features_norm, targets, test_size=0.2, random_state=42)

# Create the parameter grid
param_grid = {"max_depth": [3,4,5, None],
              "max_features": [1, 3, 5, 7, 9],
              "min_samples_leaf": [1, 3, 5, 7, 9],
              "criterion": ["mse"]}

# Instantiate the regressor
rf = RandomForestRegressor(n_estimators=100)

# Set up the grid search
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)

# Fit the grid search model
grid_search.fit(X_train, y_train)

#best estimator
best_est = grid_search.best_estimator_
best_est

# Train the model
#rf.fit(X_train, y_train)


```

This code is very similar to the RandomizedSearchCV example provided earlier. The main difference is that instead of specifying a range of values for each hyperparameter in the param_dist dictionary, we specify a set of specific values to try in the param_grid dictionary. The GridSearchCV object is set up with a RandomForestRegressor object, and the hyperparameters to be tuned are max_depth, max_features, min_samples_leaf, and criterion. The GridSearchCV object will evaluate all possible combinations of the specified parameter values and will return the best set of hyperparameters. The cv parameter controls the number of folds used in cross-validation, and the iid parameter controls whether the data is assumed to be independently and identically distributed.

Please note that using GridSearchCV will take more time as it will evaluate all the possible combinations of the hyperparameters, where as in RandomizedSearchCV it samples from the range of the hyperparameters.

```python
from sklearn.model_selection import cross_val_score
```

```python
training_score = cross_val_score(best_est, X_train, np.ravel(y_train), cv=5)
testing_score = cross_val_score(best_est, X_test, np.ravel(y_test), cv=5)

print("Classifiers: ", best_est.__class__.__name__, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy ")
print("Classifiers: ", best_est.__class__.__name__, "Has a testing score of", round(testing_score.mean(), 2) * 100, "% accuracy ")
```

```python
# Train the model
rf.fit(X_train, y_train)

training_score = cross_val_score(rf, X_train, np.ravel(y_train), cv=5)
testing_score = cross_val_score(rf, X_test, np.ravel(y_test), cv=5)

print("Classifiers: ", rf.__class__.__name__, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy ")
print("Classifiers: ", rf.__class__.__name__, "Has a testing score of", round(testing_score.mean(), 2) * 100, "% accuracy ")
```

```python
# Predict and evaluate
y_pred = best_est.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error: ", mse)
print("R-Squared: ", r2)
```
